```python
# Interactive demo - 2
# Adding subtracting constant values to/from a RGB image

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import cv2
import numpy as np

sl1 = widgets.IntSlider(value=0, min=-150, max=150, step=30);
sl2 = widgets.IntSlider(value=0, min=-150, max=150, step=30);
sl3 = widgets.IntSlider(value=0, min=-150, max=150, step=30);



def f(b_constant, g_constant, r_constant):
    img = cv2.imread(folder_path + "imgs/chapter3/castle.png");
    
    #################################FOCUS#######################################
    b, g, r = cv2.split(img)
    b = cv2.add(b, b_constant)
    g = cv2.add(g, g_constant)
    r = cv2.add(r, r_constant)
    #############################################################################
    
    img = cv2.merge((r,g,b))
    
    plt.figure(figsize = (8,8))
    plt.imshow(img)
    plt.show()

interact(f, b_constant=sl1, g_constant=sl2, r_constant=sl3);
```

### Manipluating gamma using opencv


```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread(folder_path + 'imgs/chapter3/indoor.jpg')
```

```python
# Apply Gamma=2.2 on the normalised image and then multiply by scaling constant (For 8 bit, c=255)

# Normalize image, pixels to 0-1
# Multiple by gamma
# Re-factor the image pixels to have value from 0-255

img_out = np.array(255*(img/255)**2.2,dtype='uint8')
```

```python
f = plt.figure(figsize=(15,15))
f.add_subplot(2, 1, 1).set_title('Original Image');
plt.imshow(img[:, :, ::-1])
f.add_subplot(2, 1, 2).set_title('Gamma manipulated image');
plt.imshow(img_out[:, :, ::-1])
plt.show()
```
