```python
label_id_dict = {
    "0": "Nucleoplasm",  
    "1": "Nuclear membrane",  
    "2": "Nucleoli",
    "3": "Nucleoli fibrillar center",   
    "4": "Nuclear speckles",  
    "5": "Nuclear bodies",
    "6": "Endoplasmic reticulum",   
    "7": "Golgi apparatus",   
    "8": "Peroxisomes",   
    "9": "Endosomes",   
    "10": "Lysosomes",  
    "11": "Intermediate filaments",   
    "12": "Actin filaments",   
    "13": "Focal adhesion sites",   
    "14": "Microtubules",   
    "15": "Microtubule ends",   
    "16": "Cytokinetic bridge",   
    "17": "Mitotic spindle",   
    "18": "Microtubule organizing center",   
    "19": "Centrosome",   
    "20": "Lipid droplets",   
    "21": "Plasma membrane",   
    "22": "Cell junctions",   
    "23": "Mitochondria",   
    "24": "Aggresome",   
    "25": "Cytosol",   
    "26": "Cytoplasmic bodies",   
    "27": "Rods & rings"  
}
```

```python
def map_to_classes(class_list):
    mapped_list = [];
    for i in range(len(class_list)):
        mapped_list.append(label_id_dict[class_list[i]])
    return mapped_list;
```

```python
img_name = "workspace/test/1.png"
predictions = gtf.Infer(img_name=img_name);
mapped_list = map_to_classes(predictions["predicted_classes"])
print("Classes mapped to: ", mapped_list)
from IPython.display import Image
Image(filename=img_name)
```

```python
import os
test_imgs = os.listdir("test/")
```

```python
green_test_images = [];
for i in range(len(test_imgs)):
    if("green" in test_imgs[i]):
        green_test_images.append(test_imgs[i]);
```

```python
from tqdm import tqdm

combined = [];
for i in tqdm(range(len(green_test_images))):
    predictions = gtf.Infer(img_name="test/" + green_test_images[i]);
    img_name = green_test_images[i].split(".")[0];
    wr = "";
    for j in range(len(predictions["predicted_classes"])):
        wr += predictions["predicted_classes"][j] + " ";
    wr = wr[:len(wr)-1]
    combined.append([img_name, wr]);
```

```python
import pandas as pd
df = pd.DataFrame(combined, columns = ['Id', 'Predicted'])
```

```python
df.to_csv("submission.csv", index=False);
```

# Training

```python
#Using mxnet-gluon backend 
from monk.gluon_prototype import prototype
```

```python
#Create project and experiment
ptf = prototype(verbose=1);
ptf.Prototype("Project-Protein", "sample-experiment-1");
```

```python
#Set data parameters

ptf.Dataset_Params(dataset_path="train/",
                   path_to_csv="train_labels_updated.csv",
                   
                   
                   delimiter = " ",
                   
                   
                   split=0.9,
                   input_size=224, 
                   batch_size=32, 
                   shuffle_data=True,
                   num_processors=8);
```

```python
#Apply data transformations
ptf.apply_random_horizontal_flip(train=True, val=True);
ptf.apply_normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], train=True, val=True, test=True);
```

```python
#Load dataset
ptf.Dataset();
```

```python
# Set model parameters
ptf.Model_Params(model_name="resnet18_v1", freeze_base_network=False, use_gpu=True, use_pretrained=True);
```

```python
# Load model
ptf.Model();
```

```python
# Set Training parameters
ptf.Training_Params(num_epochs=2, display_progress=True, display_progress_realtime=True, 
                    save_intermediate_models=False, 
                    intermediate_model_prefix="intermediate_model_", save_training_logs=True);
```

```python
#optimizer
ptf.optimizer_rmsprop(0.001);
```

```python
#learning rate scheduler
ptf.lr_fixed();
```

```python
#Loss function
ptf.loss_sigmoid_binary_crossentropy()
```

```python
#Start Training
ptf.Train();
```
