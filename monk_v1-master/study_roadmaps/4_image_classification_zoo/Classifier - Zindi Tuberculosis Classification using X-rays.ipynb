## Run on images for competition sumission

```python
# When installed using pip
from monk.gluon_prototype import prototype
```

```python
# Load project in inference mode

gtf = prototype(verbose=0);
gtf.Prototype("Project-TB", "Gluon-Resnet152", eval_infer=True);
```

```python
import pandas as pd
df = pd.read_csv("SampleSubmission.csv")
```

```python
from tqdm import tqdm
combined = [];
for i in tqdm(range(len(df))):
    img_name ="test_small/" + df["ID"][i] + ".png";
    predictions = gtf.Infer(img_name=img_name);
    combined.append([df["ID"][i], predictions["predicted_class"]])
```

```python
import pandas as pd
df = pd.DataFrame(combined, columns = ['ID', 'LABEL'])
```

```python
df.to_csv("submission_tb.csv", index=False)
```

# Training

```python
#Using mxnet-gluon backend 

# When installed using pip
from monk.gluon_prototype import prototype


# When installed manually (Uncomment the following)
#import os
#import sys
#sys.path.append("monk_v1/");
#sys.path.append("monk_v1/monk/");
#from monk.gluon_prototype import prototype
```

```python
gtf = prototype(verbose=1);
gtf.Prototype("Project-TB", "sample-experiment-1");
```

```python
gtf.List_Models()
```

```python
gtf.Default(dataset_path="train_small",
            path_to_csv="train_labels_modified.csv",
            model_name="resnet50_v2", 
            freeze_base_network=False,
            num_epochs=5);
```

```python
gtf.update_batch_size(16);
gtf.update_save_intermediate_models(False);

#Important to reload post updates
gtf.Reload();
```

```python
gtf.Train();
```
