```python
import os
from tqdm import tqdm
import pandas as pd
```

```python
combined = [];
img_num = 0;
major_labels = os.listdir("OfficeHomeDataset_10072016");
for i in tqdm(range(len(major_labels))):
    minor_labels = os.listdir("OfficeHomeDataset_10072016/" + major_labels[i]);
    for j in tqdm(range(len(minor_labels))):
        img_list =  os.listdir("OfficeHomeDataset_10072016/" + major_labels[i] + "/" + minor_labels[j]);
        label = major_labels[i] + "_" + minor_labels[j];
        for k in range(len(img_list)):
            cmd = "cp OfficeHomeDataset_10072016/" + major_labels[i] + "/" + minor_labels[j] + "/" + img_list[k] + " images/" + str(img_num) + ".jpg";
            combined.append([str(img_num) + ".jpg", label]);
            os.system(cmd)
            img_num += 1;
```

```python
df = pd.DataFrame(combined, columns = ['ID', 'LABEL'])  
df.to_csv("train_labels.csv", index=False);
```

## Training

```python
# Using mxnet-gluon backend 
from monk.gluon_prototype import prototype

# For pytorch backend
#from monk.pytorch_prototype import prototype

# For Keras backend
#from monk.keras_prototype import prototype
```

```python
# Create Project and Experiment

gtf = prototype(verbose=1);
gtf.Prototype("Project-Office-Home", "Gluon-vgg13_bn");
```

```python
gtf.List_Models()
```

```python
gtf.Default(dataset_path="images",
            path_to_csv="train_labels.csv",
            model_name="vgg13_bn", 
            freeze_base_network=False,
            num_epochs=100);
```

### How to change hyper parameters and models 
  - Docs - https://github.com/Tessellate-Imaging/monk_v1#4
  - Examples - https://github.com/Tessellate-Imaging/monk_v1/tree/master/study_roadmaps/1_getting_started_roadmap

```python
# Update batch size
gtf.update_batch_size(32);
gtf.update_save_intermediate_models(False);

# Reload after updates
gtf.Reload();
```

```python
#Start Training
gtf.Train();

#Read the training summary generated once you run the cell and training is completed
```
